```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

channels_nums = {
    3: [6, 16, 32],
    8: [8, 16, 32],
    16: [16, 32]
}

colors = {
    3: 'blue',
    8: 'green',
    16: 'red'
}


for fl_channel, sl_channels in channels_nums.items():
    fig = plt.figure()
    plt.ylabel("Accuracy")
    plt.xlabel("Channels of 2nd conv. layer")
    plt.title(f"Channels of 1st conv. layer: {fl_channel}")
    x = []
    y = []
    yerr = []
    for sl_channel in sl_channels:
        file_name = f"../results/net_2conv_2fc_channels_{fl_channel}_{sl_channel}.csv"
        df = pd.read_csv(file_name)
        mean = df["accuracy"].mean()
        std = df["accuracy"].std()
        x.append(sl_channel)
        y.append(mean)
        yerr.append(std*3) # three sigmas
    plt.errorbar(x, y, yerr=yerr, fmt='o', color=colors[fl_channel])


    
```
