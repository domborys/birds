```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

channels = ["8_16_32", "16_16_16", "16_32_64"]
x = []
y = []
yerr = []
for channel in channels:
    file_name = f"../results/net_3conv_2fc_channels_{channel}.csv"
    df = pd.read_csv(file_name)
    mean = df["accuracy"].mean()
    std = df["accuracy"].std()
    x.append(channel)
    y.append(mean)
    yerr.append(std*3) # three sigmas
fig = plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o')
plt.ylabel("Accuracy")
plt.xlabel("Number of channels")
plt.title("Accuracy for networks with 3 conv. layers")
    
```
