```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

learning_rates = ["0001", "001", "01"]
x = []
y = []
yerr = []
for rate in learning_rates:
    file_name = f"../results/net_2conv_2fc_optimizer_adam_{rate}.csv"
    df = pd.read_csv(file_name)
    mean = df["accuracy"].mean()
    std = df["accuracy"].std()
    rate_float = float("0."+rate)
    x.append(rate_float)
    y.append(mean)
    yerr.append(std*3) # three sigmas
fig = plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o')
plt.xscale("log")
plt.ylabel("Accuracy")
plt.xlabel("Learning rate")
plt.title("Accuracy with regard to learning rate of Adam optimizer")
    
```
