```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

nets = ["alexnet", "vgg16_bn", "resnet", "inception_v3", "mobilenet_v3_small"]
x = []
y = []
yerr = []
for net in nets:
    file_name = f"../results/{net}_pretrained.csv"
    df = pd.read_csv(file_name)
    mean = df["accuracy"].mean()
    std = df["accuracy"].std()
    x.append(net)
    y.append(mean)
    yerr.append(std)
fig = plt.figure()
plt.errorbar(x, y, yerr=yerr, fmt='o')
plt.ylabel("Accuracy")
plt.xlabel("Net architecture")
plt.title("Accuracy for different pretrained networks")
    
```
